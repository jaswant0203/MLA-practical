```python
#separating categorical and numerical columns
categorical_columns = []
numerical_columns = []

for column in data.columns:
    if (data[column].dtype == 'O'):
        categorical_columns.append(column)
    else:
        numerical_columns.append(column)
        
print(categorical_columns)
print(numerical_columns)
```

```python
def label_encoding(arr):
    uniq_cat = sorted(arr.unique())
    cat_to_num = {}
    for i in range(0, len(uniq_cat)):
        # print(i, uniq_cat[i])
        cat_to_num[uniq_cat[i]] = i     
    return cat_to_num

for i in categorical_columns:
    data[i+"_encoded"] = data[i].map(label_encoding(data[i]))
```

```python
#imported labelencoding from sklearn for converting categorical data into binary
from sklearn.preprocessing import LabelEncoder

label_encoder= LabelEncoder()
encoded_label = label_encoder.fit_transform(data["col_1"])
data["col_1_encoded"] = encoded_label


#doing this for all categorical_columns
for i in categorical_columns:
    label_encoder= LabelEncoder()
    encoded_label = label_encoder.fit_transform(data[i])
    data[i + "_encoded"] = encoded_label
```

```python
label_encoder.inverse_transform([0, 1, 1, 0])
```

```python
#drop categorical_columns after encoding
data = data.drop(categorical_columns, axis = 1)
```

## Modeling

```python
X = data.drop('col_14_encoded' ,axis=1)
Y = data['col_14_encoded']
```

```python
#splitting the dataset in training and testing
from sklearn.model_selection import train_test_split
X_train ,X_test , Y_train , Y_test = train_test_split( X ,Y ,test_size=0.3 ,random_state=42  ,stratify=Y)
```

```python
#applying naive bayes 
from sklearn.naive_bayes import GaussianNB
g = GaussianNB()
g.fit(X_train, Y_train)
```

```python

y_pred_train = g.predict(X_train)
y_pred_test = g.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print("TRAIN")
print(classification_report(Y_train, y_pred_train))
print("TEST")
print(classification_report(Y_test, y_pred_test))
```

## The accuracy we get in training dataset is 80% and in testing dataset is 79%
